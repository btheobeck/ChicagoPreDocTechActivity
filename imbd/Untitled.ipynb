```python
import scipy.io
mat = scipy.io.loadmat("imdb_crop/imdb.mat")
```

```python
print(mat)
```

```python
instances = mat["imdb"][0][0][0].shape[1]
columns = ["dob", "photo_taken","full_path", "gender","name", "face_location", "face_score", "second_face_score", "celeb_names", "celeb_id"]
import pandas as pd
df = pd.DataFrame(index = range(0,instances), columns = columns)

for i in mat:
    if i == "imdb":
        current_array = mat[i][0][0]
        for j in range(len(current_array)):
            df[columns[j]] = pd.DataFrame(current_array[j][0])
```

```python
from datetime import datetime, timedelta
df['python_datetime'] = df['dob'].apply(lambda dob: datetime.fromordinal(int(dob)) + timedelta(days=366))
```

```python
df['python_datetime'] = pd.to_datetime(df['python_datetime'], errors='coerce')
```

```python
df['year'] = df['python_datetime'].dt.year
```

```python
df['year'] = pd.to_numeric(df['year'])
```

```python
df['photo_taken'] = pd.to_numeric(df['photo_taken'])
```

```python
df.dtypes
```

```python
df['age'] = df['photo_taken'] - df['year']
```

```python
df['age']
```

```python
import numpy as np
df = df[df['face_score'] != -np.inf]
 
#some pictures include more than one face, remove them
df = df[df['second_face_score'].isna()]
 
#check threshold
df = df[df['face_score'] >= 3]
 
#some records do not have a gender information
df = df[~df['gender'].isna()]
 
df = df.drop(columns = ['name','face_score','second_face_score','face_location','python_datetime', 'celeb_names', 'photo_taken',"celeb_id" ])
```
